```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

num_rows = 50000
session_ids = list(range(1, num_rows + 1))
traffic_sources = ['web', 'mobile', 'email', 'ads', 'referral']
device_types = ['Android', 'iOS', 'Windows', 'Mac']
browsers = ['Chrome', 'Firefox', 'Edge', 'Safari', 'Opera']

today = pd.Timestamp.today()

np.random.seed(42)
random.seed(42)

traffic_source_data = np.random.choice(traffic_sources, num_rows)
device_type_data = np.random.choice(device_types, num_rows)
browser_data = [random.choice(browsers) if ts == 'web' else '' for ts in traffic_source_data]
start_times = [datetime(2025, 1, 1) + timedelta(minutes=random.randint(0, 525600)) for _ in range(num_rows)]
session_durations = [random.randint(30, 3600) for _ in range(num_rows)]
end_times = [start + timedelta(seconds=dur) for start, dur in zip(start_times, session_durations)]
is_conversion_data = np.random.choice([True, False], num_rows, p=[0.15, 0.85])

sessions = pd.DataFrame({
    'session_id': session_ids,
    'traffic_source': traffic_source_data,
    'device_type': device_type_data,
    'browser': browser_data,
    'start_time': start_times,
    'end_time': end_times,
    'session_duration': session_durations,
    'is_conversion': is_conversion_data,
    'created_at': today  # New column with today's date
})

print(sessions.info())
print(sessions.head())

sessions.to_csv('/Users/dimasadimo/Desktop/Learning/Purwadika/Final Project/datamaster/sessions.csv', index=False)
```
